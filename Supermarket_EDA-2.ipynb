```python
# Group data by 'Category' and 'Zone' and calculate sum of 'Value' for each group
grouped_data = data.groupby(['Category', 'Region'])['Profit'].sum().unstack()

# Plotting
fig, ax = plt.subplots()
# Width of each bar
bar_width = 0.2
# Position for each bar group
positions = range(len(grouped_data.index))

# Plotting bars for each zone
for i, zone in enumerate(grouped_data.columns):
    x = [p + i * bar_width for p in positions]
    ax.bar(x, grouped_data[zone], bar_width, label=zone)
# Customizing the plot
fig,ax = figsize=(15, 4)
# ax.set_xticks([p + 0.5*bar_width for p in positions])
ax.set_xticklabels(['Bakery', 'Beverages', 'Eggs, Meat & Fish', 'Food Grains','Fruits & Veggies', 'Oil & Masala', 'Snacks'])
ax.set_xlabel('Categories')
ax.set_ylabel('Region')
ax.set_title('Zone-wise Bar Plot for Various Categories')
ax.legend()
plt.show()
```
